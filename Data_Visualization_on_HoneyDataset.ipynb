# Q3) Display the percentage distribution of the data in each year using the pie chart.

```python
# calculate the percentage distribution of the data in each year
year_perdist = data.groupby('year').size()/len(data)*100

# plotting the pie chart
plt.figure(figsize=(9,9))
plt.pie(year_perdist, labels=year_perdist.index, autopct='%1.1f%%')
plt.title('Percentage Distribution of Data in Each Year')
plt.show()
```

# Q4) Plot and Understand the distribution of the variable "price per lb" using displot, and write your findings.

```python
# plotting displot of variable(price per lb)
plt.figure(figsize=(9,9))
sns.distplot(data['priceperlb'])
plt.title("distribution of the variable (price per lb) using displot")
plt.show()
```

# Q5) Plot and understand the relationship between the variables 'numcol' and 'prodval' through scatterplot, and write your findings.


```python
#plotting the scatterplot
plt.figure(figsize=(9,9))
plt.scatter(data['numcol'], data['prodvalue'],color = "tan")
plt.xlabel('Number of Colonies')
plt.ylabel('Production Value')
plt.title('Relationship between Number of Colonies and Production Value using a Scatter Plot.')
plt.show()
```

# Q6) Plot and understand the relationship between categorical variable 'year' and a numerical variable 'prodvalue' through boxplot, and write your findings.

```python
#plotting a boxplot
plt.figure(figsize=(9,9))
sns.boxplot(x='year', y='prodvalue', data=data)
plt.title('Relationship between Year and Prodvalue')
plt.show()
```

# Q7) Visualize and understand the relationship between the multiple pairs of variables throughout different years using pairplot and add your inferences.

```python
# Plotting a PairPlot
sns.pairplot(data, vars=['numcol', 'yieldpercol', 'totalprod', 'prodvalue', 'year'], hue='year')
```

# Q8) Display the correlation values using a plot and add your inferences.
(use columns 'numcol', 'yield percol','total prod', 'stocks', 'price per lb', 'prodvalue')

```python
# Create a subset of the dataframe
data_subset = data[['numcol', 'yieldpercol', 'totalprod', 'stocks', 'priceperlb', 'prodvalue']]

# Calculate correlation matrix
corr_matrix = data_subset.corr()

# Plotting the heatmap
plt.figure(figsize=(10, 8))  # Adjust the size here
sns.heatmap(corr_matrix, annot=True)
plt.title("Correlation Matrix")
plt.show()
```
